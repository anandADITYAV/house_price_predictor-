```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("housing.csv")
data_set.head()
```

```python
data_set.describe()
```

```python
data_set.isnull().sum()
```

```python
#  for removing the null data we replace in ocean_proximity NEAR BAY into 1 else 0
```

```python
data_set["ocean_proximity"].unique()
```

```python
encoded_data = pd.get_dummies(data_set['ocean_proximity'], prefix='proximity').astype(int)
```

```python
encoded_data.head()
```

```python
data_set = pd.concat([data_set, encoded_data], axis=1)
```

```python
data_set
```

```python
data_set.drop(columns=["ocean_proximity"], inplace=True)
```

```python
data_set.head()
```

```python
data_set["total_rooms"]=np.log(data_set["total_rooms"] + 1).astype(int)
data_set["population"]=np.log(data_set["population"] + 1).astype(int)
data_set["total_bedrooms"] = data_set["total_bedrooms"].fillna(0)  # Replace NaNs with 0 or another value
data_set["total_bedrooms"] = np.log(data_set["total_bedrooms"] + 1).astype(int)
```

```python
x=data_set.drop(columns=["median_house_value"])
y=data_set[["median_house_value"]]
```

```python
sns.pairplot(data_set)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=42)
```

```python
x_test.shape
```

```python
rfr=RandomForestRegressor(max_features=8,n_estimators=30)
```

```python
rfr.fit(x_train,y_train)
```

```python
rfr.score(x_test,y_test)*100
```

```python
rfr.score(x_train,y_train)*100
```

```python
y_pred = rfr.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
